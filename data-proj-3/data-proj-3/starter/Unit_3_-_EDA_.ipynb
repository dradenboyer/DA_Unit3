#### Fill in blanks with a brief description of each plot type:

  
- Histogram
> Histograms are used to show the distribution of numerical data by presenting the frequency of data values. Values are organized into bins (Ranges) to show the spread of the data. This type of visualization would be useful to check for skewness.

- Density Plot / KDE
> A density plot is like a smoother version of a histogram. Generally, the kernel density estimate is used in density plots to show the probability density function of the variable. A continuous curve, which is the kernel is drawn to generate a smooth density estimation for the whole data.
- Bar Chart
>Bar charts are useful for visualizing the count of nominal/categorical data. Bar charts are a great alternative to the issues that arise with pie charts.
- Line Plot
>A line plot visualizes data by connecting the data points via line segments
- Univariate scatter plot
>This plots different observations/values of the same variable corresponding to the index/observation number.
- Violin Plot
>The Violin plot is very much similar to a box plot, with the addition of a rotated kernel density plot on each side.
- Box Plot
>Box plots are also used to look at the spread of your data. These visualizations also give you descriptive statistics that are not featured in histograms.
- Pie Chart
>Though pie charts are highly contested in the realm of data analytics and visualization, they may offer insight for further analysis. Pie charts allow you to visualize value distribution for nominal or ordinal data.
- Rug Plot
>A rug plot is a very simple, but also an ideal legitimate, way of representing a distribution. It consists of vertical lines at each data point.

### Plot for 'age' column

```python
# create plot for age column

sns.displot(data = df, x = 'age', kind = 'hist', bins=30)

plt.xlabel("Age", fontsize=14)
plt.ylabel("Number of Managers", fontsize=14)
plt.title("Distribution of Manager Ages in Survey", fontsize=14);
```

### Plot for 'salary' column

```python
# create plot for salary column
plt.figure(figsize=(12,6))

# establishes figure with data and plot type
ax = sns.histplot(data=df, x='salary', palette="viridis", bins=8)

# sets x and y labels
plt.xlabel("Salary", fontsize=14)
plt.ylabel("Number of Managers", fontsize=14)

# sets title
plt.title("Density of Manager Salaries", fontsize=15);
```

```python
# load salary_numeric.csv into dataframe where education levels have been re-expressed as numeric
df1 = pd.read_csv('salary_numeric.csv')
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

```python
# import pearson and spearmanr methods from scipy.stats library
from scipy.stats import pearsonr, spearmanr
```

### What is the correlation between salary and education?

#### Your Hypothesis = Typically more education equals higher salary

```python
# calculate correlation and pvalue of salary and education with spearmanr method
spearmanr(df1['salary'], df1['education'])
```

[interpret result]
>.477 corelation so almost middle

#### Are these results statistically significant? How do you know?

Yes because pvalue is lower than .05

### What is the correlation between salary and age?

#### Your Hypothesis: I would say usually higher in age = higher salary

```python
# calculate correlation and pvalue of salary and age with pearsonr method
pearsonr(df1['salary'], df1['age'])
```

.663 so a little bit above average correlation betweeen age and salary

#### Are these results statistically significant? How do you know? 

Yes because pvalue is lower than .05

### What is the correlation between salary and industry experience? 

#### Your Hypothesis = More industry exp = more money

```python
# calculate correlation and pvalue of salary and industry experience with pearsonr method
spearmanr(df1['salary'], df1['industry_experience'])
```

Slightly above middle 

#### Are these results statistically significant? How do you know?

Yes because pvalue is lower than .05

#### Heatmap of crosstab of average salary grouped by education and industry

```python
# create heatmap with annotated mean salary in each industry and education level
ax = sns.heatmap(crosstab, annot=True, linewidths=.5, cmap='YlGnBu', fmt='g', xticklabels=['High School', 'Some College', "Bachelor's Degree", "Master's Degree"])
```
